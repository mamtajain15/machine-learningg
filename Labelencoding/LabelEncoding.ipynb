Label Encoding:
converting the label in to numeric form

```python
# Importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encoding of breast cancer dataset


```python
# Loading the data from csv file to panda dataframes
cancer_data = pd.read_csv('/content/data.csv')
```

```python
# first 5 rows of the dataframe
cancer_data.head()
```

```python
# finding the count of different label
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
label = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframes
cancer_data['target'] = label
```

```python
cancer_data.head()
```

0--> Benign

1--> Malignant

```python
cancer_data['target'].value_counts()
```

Label encoding of IRIS data


```python
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```
